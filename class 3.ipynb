then add the following method:

1. Append
2. `__str__`

```python
class Node:
    def __init__(self,data=None, next=None):
        self.data = data
        self.next = next
        
class LinkList:
    def __init__(self):
        self.head = None
        
    def append(self,item):
        node = Node(item)
        if not self.head:
             self.head = node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = node
            
    def __str__(self):
        out_str = "["
        current = self.head
        while current.next:
            out_str += " %s," % str(current.data)
            current = current.next
        out_str += " %s]" % str(current.data)
        return out_str
                
    
ll = LinkList()

for i in range(5):
    ll.append(i)
    
print(ll)


    
    
        
```
